## Contour plotting

Contour plots are a way of representing three-dimensional data in two dimensions. An example is shown below. The red contour lines show you points on the map with the same height. 

![contour plots](https://gisgeography.com/wp-content/uploads/2016/12/Mount-Fuji-3D-Contours-Labels-678x322.png)

![contour plots](https://gisgeography.com/wp-content/uploads/2016/12/Fuji-Mountain-Contour-2D-Label-3.png)

I'm going to give you data for a "landscape" like the one above. The matplotlib library can be used to make contour plots. 

There are two ways to do this:
1. matplotlib.pyplot.contour(**x,y,z**): shows you the contour lines
1. matplotlib.pyplot.contourf(**x,y,z**): fills in the contours

```python
import numpy as np
import matplotlib.pyplot as plt

def mountain(x, y):
    return -1000*(np.sin(x) ** 10 + np.cos(y)) + 2200

lon = np.linspace(0, 6, 50) #x - value
lat = np.linspace(0, 6, 40) #y - value
LON, LAT = np.meshgrid(lon, lat)

height_mtn = mountain(LON, LAT) #z - value

#Using matplotlib.pyplot.contour
plt.figure()
plt.contour(LON,LAT,height_mtn)
plt.show()

#Using matplotlib.pyplot.contour and matplotlib.pyplot.contourf
plt.figure()
plt.title("Contour Mapping of the Sea Floor")
ax_f = plt.contourf(LON,LAT,height_mtn)
ax = plt.contour(LON,LAT,height_mtn, colors = 'black')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def mountain(x, y):
    return -1000*(np.sin(x) ** 10 + np.cos(y)) + 2200

lon = np.linspace(0, 6, 50) #x - value
lat = np.linspace(0, 6, 40) #y - value
LON, LAT = np.meshgrid(lon, lat)

height_mtn = mountain(LON, LAT) #z - value
```

```python
#Using matplotlib.pyplot.contour
plt.figure()
plt.title("Contour Mapping of the Sea Floor")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
cm=plt.cm.get_cmap("cool")
plt.contour(LON,LAT,height_mtn,cmap=cm)
plt.colorbar(label='--------->')
plt.savefig('cool.jpg')
plt.show()
```

```python
#Using matplotlib.pyplot.contour and matplotlib.pyplot.contourf
plt.figure()
plt.title("Contour Mapping of the Sea Floor")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
cm=plt.cm.get_cmap("plasma")
ax_f = plt.contourf(LON,LAT,height_mtn,cmap=cm) #Used for background color
plt.colorbar(label='--------->')
ax = plt.contour(LON,LAT,height_mtn,colors='black') #Used for contour line color
plt.clabel(ax)
plt.savefig('plasma.jpg')
plt.show()
```

```python
#Using another color scheme ->inferno
plt.figure()
plt.title("Contour Mapping of the Sea Floor")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
cm=plt.cm.get_cmap("inferno")
ax_f = plt.contourf(LON,LAT,height_mtn,cmap=cm) #Used for background color
plt.colorbar(label='--------->')
ax = plt.contour(LON,LAT,height_mtn,colors='black') #Used for contour line color
plt.clabel(ax)
plt.savefig('inferno.jpg')
plt.show()
```
